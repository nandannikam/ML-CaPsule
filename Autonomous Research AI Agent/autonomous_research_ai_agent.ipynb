```python
import os
import getpass
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_groq import ChatGroq
from langgraph.prebuilt import create_react_agent
from langchain_core.messages import HumanMessage

# 1. Securely prompt for the Groq API Key
print("Please enter your Groq API Key (get one at https://console.groq.com/):")
os.environ["GROQ_API_KEY"] = getpass.getpass("Groq API Key:")

def run_research_agent(query):
    print(f"\n🤖 Initializing Research Agent for query: '{query}'\n")
    
    try:
        # 2. Initialize the LLM (UPDATED TO GROQ'S LATEST MODEL)
        llm = ChatGroq(model_name="llama-3.1-8b-instant")
        
        # 3. Define the Web Search Tool
        search_tool = DuckDuckGoSearchRun()
        tools = [search_tool]
        
        # 4. Create the LangGraph ReAct Agent 
        agent_executor = create_react_agent(llm, tools)
        
        # 5. Execute the Agent
        response = agent_executor.invoke({"messages": [HumanMessage(content=query)]})
        
        print("\n✅ Research Complete!")
        print("-" * 50)
        print("Agent's Final Answer:\n")
        
        # Extract the text content from the agent's final message
        final_answer = response["messages"][-1].content
        print(final_answer)
        
    except Exception as e:
        print(f"\n❌ An error occurred: {e}")

# 6. Test the agent 
sample_query = "What are the latest advancements in AI agents as of early 2024?"
run_research_agent(sample_query)
```
